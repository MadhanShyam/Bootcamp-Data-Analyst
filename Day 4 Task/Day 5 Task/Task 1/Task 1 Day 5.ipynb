```python
# 📚 Import required libraries
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# 📂 Load the dataset
file_path = 'IPL_Match_Highlights_Commentary.csv'  # Ensure this file is in the same directory
df = pd.read_csv("IPL_Match_Highlights_Commentary.csv")

# 🧹 Combine all commentary into a single text and remove stopwords
commentary_text = ' '.join(df['Commentary'].dropna().astype(str).tolist())

# 🛑 Add custom stopwords (specific to cricket context)
custom_stopwords = set(STOPWORDS)
custom_stopwords.update(['over', 'ball', 'runs', 'run', 'wicket', 'match', 'innings', 'team', 'player'])

# 🌥️ Generate the word cloud
wordcloud = WordCloud(width=1000, height=600, background_color='white',
                      stopwords=custom_stopwords, colormap='plasma',
                      contour_color='darkblue', contour_width=2).generate(commentary_text)

# 🎨 Display the word cloud (emoji removed from title)
plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('IPL Cricket Commentary Word Cloud', fontsize=18)
plt.show()
```
