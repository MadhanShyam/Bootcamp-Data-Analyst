```python
# Load the CSV file
file_path = 'Life-Expectancy-Data-Averaged.csv'
data = pd.read_csv("Life-Expectancy-Data-Averaged.csv")
```

```python
# Check the dataset's columns and data types
data.info()

# Preview the column names
print(data.columns)
```

```python
# Group by year and calculate the average life expectancy
avg_life_expectancy = data.groupby('Year')['Life expectancy'].mean().reset_index()

# Preview the averaged data
avg_life_expectancy.head()
```

```python
# Convert 'Year' to integer
data['Year'] = data['Year'].astype(int)

# Group by year and calculate the average life expectancy
avg_life_expectancy = data.groupby('Year')['Life_expectancy'].mean().reset_index()

# Preview the averaged data
avg_life_expectancy.head()
```

```python
# Check the unique years in the dataset
print(sorted(data['Year'].unique()))
```

```python
# Group by Region and calculate the average life expectancy
region_life_expectancy = data.groupby('Region')['Life_expectancy'].mean().reset_index()

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(region_life_expectancy['Region'], region_life_expectancy['Life_expectancy'], color='skyblue')

# Add title and labels
plt.title('Average Life Expectancy by Region (2007)', fontsize=14)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Average Life Expectancy', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```
