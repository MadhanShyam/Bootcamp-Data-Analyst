```python
# Step 1: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the Dataset
file_path = 'hapiscore_whr.csv'  # Update path if needed
data = pd.read_csv(file_path)

# Step 3: Reshape the Data for Plotting
melted_data = data.melt(id_vars='country', var_name='Year', value_name='Happiness_Score')
melted_data.dropna(subset=['Happiness_Score'], inplace=True)
melted_data['Year'] = pd.to_numeric(melted_data['Year'], errors='coerce')

# Step 4: Compute Average Happiness Score per Year
avg_score_per_year = melted_data.groupby('Year')['Happiness_Score'].mean().reset_index()

# Step 5: Create the Line Chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_score_per_year, x='Year', y='Happiness_Score', marker='o')
plt.title('Average Happiness Score (WHR) Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Happiness Score (WHR)', fontsize=12)
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the Dataset
file_path = 'hapiscore_whr.csv'  # Update this path if needed
data = pd.read_csv("hapiscore_whr.csv")

# Step 3: Reshape the Data for Plotting
melted_data = data.melt(id_vars='country', var_name='Year', value_name='Happiness_Score')
melted_data.dropna(subset=['Happiness_Score'], inplace=True)
melted_data['Year'] = pd.to_numeric(melted_data['Year'], errors='coerce')

# Step 4: Create the Line Chart with Legend for Different Countries
plt.figure(figsize=(16, 8))
sns.lineplot(data=melted_data, x='Year', y='Happiness_Score', hue='country', marker='o')

# Chart Details
plt.title('Happiness Score (WHR) Trends Over the Years by Country', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Happiness Score (WHR)', fontsize=14)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1)
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Filter top 5 countries by data availability
top_countries = melted_data['country'].value_counts().head(5).index.tolist()
filtered_data = melted_data[melted_data['country'].isin(top_countries)]

# Plotting
plt.figure(figsize=(14, 7))
sns.lineplot(data=filtered_data, x='Year', y='Happiness_Score', hue='country', marker='o')
plt.title('Happiness Score Trends Over the Years (Top 5 Countries)', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Happiness Score (WHR)', fontsize=14)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
